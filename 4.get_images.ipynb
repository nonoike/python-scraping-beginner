# 画像の取得

- openCV: 画像編集もできる
- Pillow: 画像の読み込みのみなどシンプルな操作の場合は楽に扱える

```python
from selenium import webdriver
browser = webdriver.Chrome()
browser.get('https://scraping-for-beginner.herokuapp.com/image')
```

```python
!mkdir image
```

## 個別取得

```python
from PIL import Image
```

```python
elem = browser.find_element_by_class_name('material-placeholder').find_element_by_tag_name('img')
url = elem.get_attribute('src')
```

```python
import io
from urllib import request
```

```python
f = io.BytesIO(request.urlopen(url).read())
f
```

```python
img = Image.open(f)
img
```

```python
img.save('img01.jpg')
```

## 一括取得

```python
elems = browser.find_elements_by_class_name('material-placeholder')
for elem in elems:
    url = elem.find_element_by_tag_name('img').get_attribute('src')
    f = io.BytesIO(request.urlopen(url).read())
    img = Image.open(f)
    
    # img.resize(600, 400)
    filename = url.split('/')[-1]
    img.save('image/{}'.format(filename))
```
